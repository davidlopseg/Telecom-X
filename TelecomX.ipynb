```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')

# El archivo 'TelecomX_Data.json' está en la raíz de 'Mi Unidad' (My Drive).
file_path = '/content/drive/MyDrive/TelecomX_Data.json'
df = pd.read_json(file_path)

display(df.head())
```

### Aplanamiento de la Columna 'Charges'

La columna `Charges` aún contiene diccionarios anidados para los cargos mensuales y totales. Vamos a extraer estos valores en nuevas columnas (`MonthlyCharges` y `TotalCharges`) y convertirlos a tipos de datos numéricos para su correcto análisis.

```python
# Aplanamiento de la columna 'Charges'
charges_df = df['Charges'].apply(pd.Series)
df = pd.concat([df, charges_df], axis=1)
df = df.drop('Charges', axis=1)

# Renombrar las nuevas columnas para mayor claridad
df = df.rename(columns={'Monthly': 'MonthlyCharges', 'Total': 'TotalCharges'})

# Verificar tipos de datos y manejar posibles errores de conversión
# La columna 'TotalCharges' a menudo contiene espacios en blanco o valores no numéricos que deben manejarse.
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

print("DataFrame después de aplanar y convertir la columna 'Charges':")
display(df.head())
print("Tipos de datos actualizados para MonthlyCharges y TotalCharges:")
print(df[['MonthlyCharges', 'TotalCharges']].dtypes)
```

### One-Hot Encoding de Columnas Categóricas

### Estandarización y Transformación de Datos

```python
# Mapear 'Yes' a 1 y 'No' a 0 para columnas binarias simples
YES_NO_MAP = {'Yes': 1, 'No': 0}

# Columnas con 'Yes'/'No'
binary_cols = [
    'Partner', 'Dependents', 'PaperlessBilling'
]

for col in binary_cols:
    df[col] = df[col].map(YES_NO_MAP)

# Mapear 'Female' a 1 y 'Male' a 0 para la columna 'gender'
df['gender'] = df['gender'].map({'Female': 1, 'Male': 0})

# Mapear 'Yes' a 1, 'No' a 0 y 'No phone service' a 0 para 'PhoneService' y 'MultipleLines'
df['PhoneService'] = df['PhoneService'].map(YES_NO_MAP)
df['MultipleLines'] = df['MultipleLines'].map({'Yes': 1, 'No': 0, 'No phone service': 0})

# Mapear 'Yes' a 1, 'No' a 0 y 'No internet service' a 0 para los servicios de internet
internet_service_cols = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]
INTERNET_SERVICE_MAP = {'Yes': 1, 'No': 0, 'No internet service': 0}

for col in internet_service_cols:
    df[col] = df[col].map(INTERNET_SERVICE_MAP)

# Convertir 'Churn' a formato binario (1 si 'Yes', 0 si 'No')
df['Churn'] = df['Churn'].map(YES_NO_MAP)

print("DataFrame después de la estandarización y transformación de columnas binarias:")
display(df.head())
print("Tipos de datos actualizados después de la transformación:")
print(df[binary_cols + ['gender', 'PhoneService', 'MultipleLines'] + internet_service_cols + ['Churn']].dtypes)
```

Ahora que las columnas binarias están transformadas, podemos inspeccionar el resto de las columnas categóricas (objetos) para ver cuáles requieren otro tipo de codificación (como One-Hot Encoding) y cuáles ya están listas.

```python
print("Valores únicos de las columnas categóricas restantes:")
for column in df.select_dtypes(include='object').columns:
    print(f"\nColumna '{column}':")
    print(df[column].unique())
```
